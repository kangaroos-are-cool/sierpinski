```python
import matplotlib.pyplot as plt
import math
import random

ratio = 0.50
nodes = 3
rounds = 100000

r = 1
x = []
y = []

delta = (2*math.pi)/nodes 


#setting starting positions for nodes
for i in range(nodes):
    x_i = r*math.cos(delta*i + (math.pi/2))
    y_i = r*math.sin(delta*i + (math.pi/2))
    x.append(x_i)
    y.append(y_i)

# starting position
current_x = 0.0
current_y = 0.0

point_x = [current_x]
point_y = [current_y]

for i in range(rounds):
    # Randomly Selecting a Node
    ran = random.randint(0, nodes-1)
    mag = math.sqrt((x[ran] - current_x)**2 + (y[ran] - current_y)**2)
    ux = (x[ran]-current_x)/mag
    uy = (y[ran]-current_y)/mag
    mag = mag*ratio
    point_x.append(ux*mag)
    point_y.append(uy*mag)
    current_x = (current_x-x[ran])*ratio
    current_y = (current_y-y[ran])*ratio

plt.scatter(point_x, point_y, 0.01)
plt.scatter(x, y) # display position of original nodes
plt.show()
```
